```python
import pandas as pd
```

```python
voos = pd.read_csv('dados/tabelasParaJuncao/voos.csv', header=0, sep= ';', encoding= 'utf-8')
voos.head()
```

```python
qntVoos = pd.read_csv('dados/tabelasParaJuncao/quatidade_de_voos.csv', header = 0, sep = ';', encoding = 'utf-8')
qntVoos.head()
```

```python
pd.merge(voos, qntVoos, on='month')
```

```python
produtos1 = pd.read_csv('dados/tabelasParaJuncao/produtos1.csv', decimal=',', header=0, sep= ';', encoding= 'utf-8', index_col=['id'])
produtos1.head(10)
```

```python
produtos2 = pd.read_csv('dados/tabelasParaJuncao/produtos2.csv',decimal=',', header=0, sep= ';', encoding= 'utf-8', index_col=['id'])
produtos2.head(10)
```

```python
produtos = pd.concat([produtos1, produtos2])
produtos.head(10)
```

```python
vendas = pd.read_csv('dados/tabelasParaJuncao/vendas.csv', header=0, sep= ';', encoding= 'utf-8')
vendas.head(10)
```

```python
vendas = vendas.join(produtos, on='produto_id')
vendas.head()
```

```python
vendas['total'] = vendas.quantidade * vendas.preco
```
