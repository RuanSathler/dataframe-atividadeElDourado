```python
import pandas as pd
```

```python
titanic = pd.read_csv('dados/titanic.csv', header=0, sep= ';', encoding= 'utf-8')
titanic
```

```python
sex = pd.get_dummies(titanic.sex)
sex
```

```python
titanic[['female', 'male']] = sex
titanic.head()
```

```python
titanic.drop(['sex'],axis=1, inplace = True)
titanic.drop(['male'], axis=1, inplace = True)
titanic.head()
```

```python
titanic.loc[(titanic.female == 0) &(titanic.who == 'whoman')] | titanic.loc[(titanic.female == 1) &(titanic.who == 'man')]
```

```python
titanic.loc[(titanic.survived == 0) &(titanic.alive == 'yes')] | titanic.loc[(titanic.survived == 0) &(titanic.who == 'no')]
```

```python
titanic.drop(['who', 'alive'], axis=1, inplace= True)
titanic
```
